## Reading JSON data

Pandas has a built in function to read json data that we can use to directly read from the API. We just need the api key and a specific link to have the api grab data from. By not hardcoding these into the dataframe we are able to change api keys and more importantly easily change links. In this case we are utilizing the comments API.

```python
import pandas as pd

apiKey = "Z0WG1iCCsoNRXMsEUv9vNGeDD14jGGbH"
link = "https://www.nytimes.com/2020/04/13/business/media/joe-biden-tara-reade-new-york-times.html"
df = pd.read_json(f"https://api.nytimes.com/svc/community/v3/user-content/url.json?api-key={apiKey}&offset=0&url={link}")
```

## The DataFrame

Without any filtering or modification done to the dataframe it isn't terribly useful beyond telling us the total amount of comments and other aggregate data. 

```python
df
```

## Filtering the DataFrame

Say we wanted more data about individual comments such as what comments are made by who and from where. Since, we know the format of the JSON data we are able to filter the dataframe down to just the specific columns that we want.

```python
comments = pd.DataFrame(df["results"]["comments"]).filter(["commentID","userDisplayName","userLocation","commentBody"]).head()
comments
```
